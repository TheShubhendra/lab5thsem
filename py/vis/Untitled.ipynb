```python

x = np.linspace(0, np.pi, 50)

plt.ion()
fig, ax = plt.subplots()

graph = ax.plot(x,np.sin(x),color = 'g')[0]
plt.pause(0.1)
graph = ax.plot(x,np.sin(x+1),color = 'g')[0]
# plt.pause(1)
```

```python
np.random.seed(19) 
data = np.random.random((5, 10, 10)) 
  
   
for i in range(len(data)): 
    plt.cla() 
    plt.title('matplotlib.pyplot.pause() function Example\n\n Window {}'.format(i), fontweight ="bold") 
    plt.imshow(data[i]) 
    plt.pause(0.1) 
```
