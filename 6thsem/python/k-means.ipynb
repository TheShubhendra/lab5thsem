```python
from sklearn.datasets import load_iris
from sklearn.cluster import KMeans
```

```python
dataset = load_iris()
```

```python
model = KMeans(n_clusters=3)
clustered = model.fit_transform(dataset.data)
```

```python
model.n_clusters
```

```python
from sklearn.manifold import TSNE
```

```python
tsne
```

```python
tsne = TSNE(n_components=3)
reduced = tsne.fit_transform(dataset.data)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(reduced[:, 0], reduced[:, 1], c=dataset.target)
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(reduced[:, 0], reduced[:, 1], reduced[:, 2], c=dataset.target)
```

```python
clustered.shape
```

```python
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(clustered[:, 0], clustered[:, 1], clustered[:, 2], c=dataset.target)
```
